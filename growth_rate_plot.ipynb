```python
import numpy as np
import matplotlib.pyplot as plt
kbins=223

k = np.genfromtxt("pwspctrm.txt",usecols=0,delimiter="\t")
power = np.genfromtxt("pwspctrm.txt",usecols=1,delimiter="\t")

k_epochs = [k[i*kbins:(i+1)*kbins] for i in range(int(len(k)/kbins))] 
power_epochs = [power[i*kbins:(i+1)*kbins] for i in range(int(len(k)/kbins))] 

k_epochs = np.array(k_epochs)
power_epochs = np.array(power_epochs)

print(len(k_epochs),len(power_epochs))
```

```python
growth = power_epochs[-1,20:200]/power_epochs[1,20:200]
plt.plot(k_epochs[0,20:200],np.sqrt(growth))
```
